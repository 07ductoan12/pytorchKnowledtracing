```python
import pandas as pd
import numpy as np
```

```python
df = (
    pd.read_csv(
        "./datasets/ASSIST2009/skill_builder_data.csv",
        encoding="Windows-1252",
        low_memory=False,
    )
    .dropna(subset=["skill_name"])
    .drop_duplicates(subset=["order_id", "skill_name"])
    .sort_values(by=["order_id"])
)
df
```

```python
u_list = np.unique(df["user_id"].values)
u_list, len(u_list)
```

```python
q_list = np.unique(df["skill_name"].values)
q_list, len(q_list)
```

```python
u2idx = {u: idx for idx, u in enumerate(u_list)}
q2idx = {q:idx for idx, q in enumerate(q_list)}
u2idx, q2idx
```

```python
q_seqs = []
r_seqs = []
ms_seqs = []
```

```python
for u in u_list:
    df_u = df[df["user_id"] == u]
    q_seq = np.array([q2idx[q] for q in df_u["skill_name"]])
    r_seq = df_u["correct"].values
    ms_seq = df_u["ms_first_response"].values
    r_seqs.append(r_seq)
    q_seqs.append(q_seq)
    ms_seqs.append(ms_seq)
```

```python
q_seqs[0], r_seqs[0], ms_seqs[0]
```

```python
proc_q_seqs = []
proc_r_seqs = []
seq_len = 100
```

```python
for q_seq, r_seq in zip(q_seqs, r_seqs):
    i = 0
    print(q_seq)
    print(r_seq)
    while i + seq_len + 1 < len(q_seq):
        proc_q_seqs.append(q_seq[i : i + seq_len + 1])
        proc_r_seqs.append(r_seq[i : i + seq_len + 1])
        i += seq_len + 1
    proc_q_seqs.append(
        np.concatenate([q_seq[i:], np.array([-1] * (i + seq_len + 1 - len(q_seq)))])
    )
    proc_r_seqs.append(
        np.concatenate([r_seq[i:], np.array([-1] * (i + seq_len + 1 - len(q_seq)))])
    )
    break

proc_q_seqs, proc_r_seqs
```
