```python
df_filtered = df_all[(df_all["correct"] == 0.0) | (df_all["correct"] == 1.0)]
```

```python
len(df_filtered)

users = list(df_filtered["user_id"])
users = list(set(users)) # unique user
print(len(users))

questions = list(df_filtered["sequence_id"])
questions = list(set(questions)) # unique question (KC)
print(len(questions))
```

```python
df_filtered = df_filtered.drop(columns=["log_id"])
df_filtered.iloc[:10]
```

```python
corrects = list(df_filtered["correct"])
corrects = list(map(int, corrects))
df_filtered["correct"] = corrects
df_filtered.iloc[:10]
```

```python
questions.sort()
qnum = len(questions)
snum = qnum
qmap = {}
for i in range(qnum):
    qmap[questions[i]] = i+1
```

```python
qs = list(df_filtered["sequence_id"])
qs = [qmap[x] for x in qs]
qs[:10]
```

```python
df_filtered["sequence_id"] = qs
df_filtered.iloc[:10]
```

```python
with open("qid_to_sid.csv", "w") as f:
    for i in range(1, qnum+1):
        f.write(f"{i},{i}\n")

with open("sid_to_qid.csv", "w") as f:
    for i in range(1, qnum+1):
        f.write(f"{i},{i}\n")
```

```python
qid_tot_cnt = {k: 0 for k in range(1, qnum + 1)}
qid_cor_cnt = {k: 0 for k in range(1, qnum + 1)}

for q, c in zip(qs, corrects):
    qid_tot_cnt[q] += 1
    if c == 1:
        qid_cor_cnt[q] += 1
```

```python
qid_gc = {k:0.0 for k in range(1, qnum+1)}
for q in range(1, qnum+1):
    qid_gc[q] = qid_cor_cnt[q] / qid_tot_cnt[q]
```

```python
qid_gc
```

```python
with open("qid_to_gc_cnt.csv", "w") as f:
    for q in range(1, qnum + 1):
        f.write(f"{q},{qid_gc[q]},{qid_tot_cnt[q]}\n")

with open("sid_to_gc_cnt.csv", "w") as f:
    for q in range(1, qnum + 1):
        f.write(f"{q},{qid_gc[q]},{qid_tot_cnt[q]}\n")
```

```python
df_filtered = df_filtered.rename(columns={"sequence_id": "q_id", "correct": "correctness"})
df_filtered.iloc[:10]
```

```python
gk = df_filtered.groupby(["user_id"])
```

```python
gk.get_group(50121)
```

```python
users = list(df_filtered["user_id"])
users = list(set(users))
len(users)
```

```python
cv_num = 5
train_rate = 0.8
test_rate = 0.2
num_users = len(users)

random.seed(0)
random.shuffle(users)

test_users = users[int(train_rate * num_users) :]
tv_users = users[: int(train_rate * num_users)]
print(len(test_users))

with open("test_users.csv", "w") as f:
    for u in test_users:
        f.write(f"{u}.csv\n")
num_tv_users = len(tv_users)

for i in range(5):
    train_users = (
        tv_users[: int(i * num_tv_users / cv_num)]
        + tv_users[int((i + 1) * num_tv_users / cv_num) :]
    )
    val_users = tv_users[
        int(i * num_tv_users / cv_num) : int((i + 1) * num_tv_users / cv_num)
    ]
    print(len(train_users), len(val_users))

    with open(f"train_users_{i+1}.csv", "w") as f:
        for u in train_users:
            f.write(f"{u}.csv\n")
    with open(f"val_users_{i+1}.csv", "w") as f:
        for u in val_users:
            f.write(f"{u}.csv\n")
```

```python
if not os.path.exists("users"):
    os.mkdir("users")
    
for u in users[:]:
    df_user = gk.get_group(u)
    df_user = df_user.drop(columns=["user_id"])
    user_path = os.path.join("users", f"{u}.csv")
    df_user.to_csv(user_path, header=True, index=False)
```
